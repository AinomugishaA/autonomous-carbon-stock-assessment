```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# I ingest the live drone telemetry file generated by my PySimverse simulation
df = pd.read_csv("forest_carbon_assessment.csv")

# I establish the environmental economic baseline variables based on standard carbon market metrics
CARBON_STOCK_PER_ZONE = 250   # Estimated metric tons of CO2 locked in a healthy forest zone
CARBON_MARKET_PRICE = 45.00   # Global voluntary market price per ton (USD)

# I apply resource economics formulas to calculate environmental damage and financial liability
df['co2_tons_lost'] = (df['canopy_loss_pct'] / 100) * CARBON_STOCK_PER_ZONE
df['economic_loss_usd'] = df['co2_tons_lost'] * CARBON_MARKET_PRICE

# I generate the aggregate summary metrics to shape my final portfolio report narrative
total_financial_loss = df['economic_loss_usd'].sum()
avg_deforestation = df['canopy_loss_pct'].mean()

print(f"====================================================")
print(f"🌲 MY ENVIRONMENTAL ECONOMIC DATA REPORT")
print(f"====================================================")
print(f"Average Forest Canopy Degradation: {avg_deforestation:.2f}%")
print(f"Total Monetized Carbon Stock Liability: ${total_financial_loss:,.2f} USD\n")

# I engineer my custom portfolio visualization for presentation
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 5))
barplot = sns.barplot(x="zone_id", y="economic_loss_usd", data=df, palette="Reds_d")

plt.title('Carbon Stock Financial Damage Assessment via Automated Drone Metrics', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Scanned Grid Management Zone ID', fontsize=12, labelpad=10)
plt.ylabel('Simulated Economic Loss (USD)', fontsize=12, labelpad=10)

# I overlay computed financial asset tags directly onto the geometric bars for better readability
for p in barplot.patches:
    barplot.annotate(f"${p.get_height():,.2f}", 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha='center', va='center', xytext=(0, 8), textcoords='offset points', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```
